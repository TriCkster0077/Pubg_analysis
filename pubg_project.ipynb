````python
plt.hist(pubg_gold_df.Matches_Played, alpha = 0.9, bins = 4)
plt.hist(pubg_platinum_df.Matches_Played, alpha = 0.4, bins = 4)
plt.hist(pubg_diamond_df.Matches_Played, alpha = 0.4, bins = 4)
plt.hist(pubg_silver_df.Matches_Played, alpha = 0.4, bins = 4)
plt.legend(['Gold', 'Platinum', 'Diamond', 'Silver']);
````

**TODO** - Using barplot for more analysis

````python
sns.barplot(x = 'Rank', y = 'Matches_Played', data = pubg_df)
````

````python
fig, axes = plt.subplots(2, 3, figsize=(19,17))

#Ranks distribution
axes[0,0].set_title('People in different Ranks')
axes[0,0].hist(pubg_df.Rank);

#Kills vs KD
axes[1,2].set_title('Kills vs KD')
sns.scatterplot(x = pubg_df.Kills, y = pubg_df.KD, hue = pubg_df.Rank, ax = axes[1,2]);

#Matches_Played vs KD
axes[0,1].set_title('KD on Matches')
sns.scatterplot(x = pubg_df.Matches_Played, y = pubg_df.KD, hue = pubg_df.Rank, ax = axes[0,1]);

#Histogram of Ranks on basis of KD
axes[0,2].set_title('Ranks on KD')
axes[0,2].hist([pubg_gold_df.KD, pubg_platinum_df.KD, pubg_diamond_df.KD, pubg_silver_df.KD])
axes[0,2].legend(['Gold', 'Platinum', 'Diamond', 'Silver']);

#Overlapping Histogram representing Matches Played
axes[1,0].set_title('Matches Played on Ranks')
axes[1,0].hist(pubg_gold_df.Matches_Played, alpha = 0.5, bins = 4)
axes[1,0].hist(pubg_platinum_df.Matches_Played, alpha = 0.5, bins = 4)
axes[1,0].hist(pubg_diamond_df.Matches_Played, alpha = 0.5, bins = 4)
axes[1,0].hist(pubg_silver_df.Matches_Played, alpha = 0.5, bins = 4)
axes[1,0].legend(['Gold', 'Platinum', 'Diamond', 'Silver']);

#Barplot on Matches based on Ranks
axes[1,1].set_title('Matches Played On Ranks')
sns.barplot(x = 'Rank', y = 'Matches_Played', data = pubg_df, ax = axes[1,1]);
````
